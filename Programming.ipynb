## Code et algorithmes sur Python

```python
def anagramme (string1, string2) :
    if (sorted(string1) == sorted(string2)):
        return("Anagrammes")
    else :
        return ("Pas anagrammes") 

print(anagramme("chien", "niche"))
print(anagramme("chien", "chat"))
```

```python
def facto(x) :
    f = 1
    if x < 0 :
        return("Choisir un autre nombre")
    elif x == 0 :
        return(1)
    else :
        for i in range(1, x+1):
            f = f*i
    return(f)

print(facto(-4))
print(facto(0))
print(facto(4))
```

```python
def triangle(x):
    n = 1
    for i in range(x):
        for j in range(i+1):
            print(n, end = " ")
            n = n + 1
        print("\n")

triangle(4)
```
